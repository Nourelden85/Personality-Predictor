```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix


from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import roc_auc_score
```

```python
x_train = pd.read_csv("../data/processed/x_train.csv")
x_test = pd.read_csv("../data/processed/x_test.csv")

y_train = pd.read_csv("../data/processed/y_train.csv")
y_test = pd.read_csv("../data/processed/y_test.csv")
```

## **Model Training**

### Logistic Regression with Grid Search 

```python
log_reg = LogisticRegression(max_iter = 1000 , random_state=42)

param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 50, 100],
    'class_weight': [None, 'balanced']
}


grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

```

```python
models = {}
for trait in y_train.columns:
    grid_search.fit(x_train, y_train[trait])
    best_model = grid_search.best_estimator_
    models[trait] = best_model


for trait, model in models.items():

    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
   
    y_pred = model.predict(x_test)
    print(f"Classification Report for {trait}:\n")
    print(classification_report(y_test[trait], y_pred))
    print(f"Confusion Matrix for {trait}:\n")
    print(confusion_matrix(y_test[trait], y_pred))
    print("Train Accuracy:", accuracy_score(y_train[trait], y_pred_train))
    print("Test Accuracy:", accuracy_score(y_test[trait], y_pred_test))
    print('\n')
```

### Decision Tree with Grid Search 

```python
param_grid = {
    'max_depth': [3,5,7,10],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,5]
}

grid = GridSearchCV(DecisionTreeClassifier(random_state=42),
                    param_grid,
                    cv=5,
                    scoring='accuracy')

```

```python
models = {}
for trait in y_train.columns:
    grid.fit(x_train, y_train[trait])
    best_dt = grid.best_estimator_
    models[trait] = best_dt


for trait, model in models.items():

    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
   
    y_pred = model.predict(x_test)
    print(f"Classification Report for {trait}:\n")
    print(classification_report(y_test[trait], y_pred))
    print(f"Confusion Matrix for {trait}:\n")
    print(confusion_matrix(y_test[trait], y_pred))
    print("Train Accuracy:", accuracy_score(y_train[trait], y_pred_train))
    print("Test Accuracy:", accuracy_score(y_test[trait], y_pred_test))
    print('\n')
```
